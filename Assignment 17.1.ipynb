```python
import pandas as pd
from scipy import stats
titanic = ('C:\\Users\\sowmy\\Downloads\\titanic-train.csv')
titanic_df= pd.read_csv(titanic)
print(titanic_df)

titanic_df['Survived'].mean()
```

```python
titanic_df.groupby('Pclass').mean()
```

```python

class_sex_grouping = titanic_df.groupby(['Pclass','Sex']).mean()
class_sex_grouping
```

```python
class_sex_grouping['Survived'].plot.bar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
group_by_age = pd.cut(titanic_df["Age"], np.arange(0, 90, 10))
age_grouping = titanic_df.groupby(group_by_age).mean()
age_grouping['Survived'].plot.bar()
plt.show()
```

```python
titanic_df.groupby('Parch').mean()
class_Parch_grouping = titanic_df.groupby(['Parch']).mean()
class_Parch_grouping

class_Parch_grouping['Survived'].plot.bar()
plt.show()
```

```python
titanic_df.groupby('SibSp').mean()
class_SibSp_grouping = titanic_df.groupby(['SibSp']).mean()
class_SibSp_grouping

class_SibSp_grouping['Survived'].plot.bar()
plt.show()
```

```python
import seaborn as sns
facet = sns.FacetGrid(titanic_df, hue="Survived",aspect=4)
facet.map(sns.kdeplot,'Age',shade= True)
facet.set(xlim=(0, titanic_df['Age'].max()))
facet.add_legend()
```

```python
import seaborn as sns
fig, axis1 = plt.subplots(1,1,figsize=(18,4))
average_age = titanic_df[["Age", "Survived"]].groupby(['Age'],as_index=False).mean()
sns.barplot(x='Age', y='Survived', data=average_age)
plt.show()
```
